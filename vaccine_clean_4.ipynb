```python
df.loc[(df['vaccine_01']=='Sputnic')|(df['vaccine_01']=='Astrazeneca'),'vac_1']=1
df.loc[df['vaccine_01'].isna()==True,'vac_1']=0
df['vac_1'].fillna(2,inplace=True)

df.loc[(df['vaccine_02']=='Sputnic')|(df['vaccine_02']=='Astrazeneca'),'vac_2']=1
df.loc[df['vaccine_02'].isna()==True,'vac_2']=0
df['vac_2'].fillna(2,inplace=True)

df.loc[(df['vaccine_03']=='Sputnic')|(df['vaccine_03']=='Astrazeneca'),'vac_3']=1
df.loc[df['vaccine_03'].isna()==True,'vac_3']=0
df['vac_3'].fillna(2,inplace=True)
```
